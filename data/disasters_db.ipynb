```python
import sqlite3
import json
import os
```

```python
with open('data_partner_api.json', 'r') as json_file:
    data = json.load(json_file)
```

```python
#Creating the incidents database

conn = sqlite3.connect('incidents.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS incidents (
    id INTEGER PRIMARY KEY,
    latitude REAL,
    longitude REAL,
    severity TEXT,
    incident_type TEXT,
    time TEXT
)
''')

for incident in data:
    cursor.execute('''
    INSERT INTO incidents (id, latitude, longitude, severity, incident_type, time)
    VALUES (?, ?, ?, ?, ?, ?)
    ''', (incident['id'], incident['latitude'], incident['longitude'], incident['severity'], incident['incident_type'], incident['time']))

conn.commit()
conn.close()
```

```python
#Visualizing the incidents database
conn = sqlite3.connect('incidents.db')
cursor = conn.cursor()

cursor.execute('SELECT * FROM incidents')

rows = cursor.fetchall()

for row in rows:
    print(row)
```
